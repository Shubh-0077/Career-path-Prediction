### Exploratory Data Analysis (EDA) - Key Findings

#### Career Role Distribution

The distribution of the target variable (*Suggested Job Role*) was analyzed to understand the balance among career categories. The dataset contains 12 distinct career roles, with Network Security Engineer having the highest number of records and Mobile Applications Developer having the lowest. However, the difference in frequencies across categories is relatively small, indicating that the dataset is well-balanced and does not suffer from significant class imbalance. This balanced distribution is beneficial for machine learning model training as it reduces prediction bias toward any particular career role.

#### Correlation Analysis of Numerical Features

A correlation analysis was performed on the numerical features, including Logical Quotient Rating, Hackathons, Coding Skills Rating, and Public Speaking Points. The correlation matrix revealed extremely low correlation values among all numerical variables. This indicates that each feature captures a different aspect of a student's profile and that there is minimal redundancy among the numerical predictors. The absence of strong correlations also suggests that multicollinearity is not a concern for the modeling phase.

#### Coding Skills Rating Across Career Roles

The distribution of Coding Skills Rating was examined across all career categories using boxplots. The analysis showed that most career roles share similar coding skill distributions, with median values centered around the middle range. No single career role demonstrated a significantly distinct coding skill profile. This suggests that coding skills alone are insufficient for determining a student's recommended career path and that multiple factors contribute to the final career prediction.

#### Management vs Technical Preference Analysis

A comparison between Management and Technical preferences across career roles showed a nearly balanced distribution. Both preference types are represented across all career categories, indicating that management orientation alone is not a dominant factor in determining career recommendations. The machine learning model will therefore need to consider combinations of features, including interests, certifications, workshops, and personality traits, to accurately predict career outcomes.

#### Overall Conclusion

The exploratory analysis indicates that the dataset is clean, balanced, and suitable for multi-class classification. Numerical features exhibit low correlation, suggesting independent information contribution, while career recommendations appear to be influenced by combinations of categorical and behavioral attributes rather than any single feature. These findings support the use of advanced machine learning algorithms capable of capturing complex relationships between student characteristics and career outcomes.

## Phase 4: Feature Engineering

```python
# Create Features and Target 
X = df.drop("suggested_job_role", axis=1)
y = df["suggested_job_role"]
```

```python
print(X.shape) 
print(y.shape)
```

```python
# Identify Numerical and Categorical Columns
cat_cols = X.select_dtypes(include="object").columns
num_cols = X.select_dtypes(exclude="object").columns

print("Categorical Features:", len(cat_cols))
print("Numerical Features:", len(num_cols))
```

```python
# Encode Categorical Features 

# One-Hot Encoding
X_encoded = pd.get_dummies(
    X,
    drop_first=True) 

X_encoded.shape
```

```python
# Encode Target Variable 
from sklearn.preprocessing import LabelEncoder

target_encoder = LabelEncoder()
y_encoded = target_encoder.fit_transform(y)
```

```python
target_encoder.classes_
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)

print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

### Feature Engineering Summary

- Features and target variable were separated for model development.
- The dataset contains 15 categorical features and 4 numerical features.
- One-Hot Encoding was applied to convert categorical variables into machine-readable numerical features.
- The feature space expanded from 19 original features to 83 encoded features.
- The target variable was encoded using LabelEncoder.
- The dataset was split into training (80%) and testing (20%) subsets.
- Stratified sampling was used to maintain class balance across training and testing datasets.
- The processed dataset is ready for machine learning model training and evaluation.

## Phase 5B — Feature Selection 

```python
# Mutual Information

from sklearn.feature_selection import mutual_info_classif

mi_scores = mutual_info_classif(
    X_encoded,
    y_encoded,
    random_state=42
)

mi_df = pd.DataFrame({
    "Feature": X_encoded.columns,
    "MI Score": mi_scores
})

mi_df = mi_df.sort_values(
    by="MI Score",
    ascending=False
)

mi_df.head(20)
```

```python
# Visualize Top 20
top20_mi = mi_df.head(20)

plt.figure(figsize=(10,6))

sns.barplot(
    data=top20_mi,
    x="MI Score",
    y="Feature"
)

plt.title("Top 20 Features by Mutual Information")

plt.tight_layout()
plt.show()
```

```python
# Chi-Square Feature Selection
from sklearn.feature_selection import chi2

chi_scores, p_values = chi2(
    X_encoded,
    y_encoded
)

chi_df = pd.DataFrame({
    "Feature": X_encoded.columns,
    "Chi2 Score": chi_scores,
    "P Value": p_values
})

chi_df = chi_df.sort_values(
    by="Chi2 Score",
    ascending=False
)

chi_df.head(20)
```

### Feature Selection Findings

- Feature selection was performed using Mutual Information and Chi-Square analysis.
- Public speaking points, logical quotient rating, hackathons, certifications, and some interest-related attributes consistently ranked among the most informative features.
- Mutual Information scores were generally low, indicating weak relationships between individual features and the target variable.
- No single feature demonstrated strong predictive power for career role classification.
- The findings suggest that the target variable may depend on complex interactions among multiple features rather than any single attribute.
